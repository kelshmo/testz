# creating 4 data frames 

```python
a='HSB'
list_a = []
for num in range(100, 200,1):list_a.append(a+str(num))
list_b = []
for num in range(100, 200,1):list_b.append(a+str(num))
list_c = []
for num in range(100, 300,2):list_c.append(a+str(num))
```

```python
mean = 0
sd = 1
n = 100 
li = []
df1 = pd.DataFrame({'individualID': list_a,'a': np.random.normal (mean, sd, n), 
                        'b': np.random.normal(mean, sd, n),'c': np.random.normal(mean, sd, n)})
df2 = pd.DataFrame({'individualID': list_b,'d': np.random.normal (mean, sd, n), 
                        'e': np.random.normal(mean, sd, n),'f': np.random.normal(mean, sd, n)})
df3 = pd.DataFrame({'individualID': list_a,'z': np.random.normal (mean, sd, n), 
                        'g': np.random.normal(mean, sd, n),'h': np.random.normal(mean, sd, n),'w': np.random.normal(mean, sd, n)})
df4 = pd.DataFrame({'individualID': list_c,'d': np.random.normal (mean, sd, n), 
                        'i': np.random.normal(mean, sd, n),'j': np.random.normal(mean, sd, n)})
```

# Visual representation: Left join
* Output Table will have only the matchingrows and columns from all tablesisual representation: LEFT join
* all the rows on table A will exist in the output table
* not all rows from table B will be in the output table
* table A and output table should have same num of rows
* if the output table does not have same amount of rows it could be duplicated data

```python
path = '/Users/mhevia/gitHub2/testz/LEFT-JOIN.png'
Image(filename=path, width=5000, height=500)
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
```

```python
df_l = pd.merge(df1, df2, how='left', left_on=['individualID'], right_on=['individualID'])
df_l.head()
```

```python
df_ll = pd.merge(df3, df4, how='left', left_on=['individualID'], right_on=['individualID'])
df_ll.head()
```

```python
df_left = pd.merge(df_l, df_ll, how='left', left_on=['individualID'], right_on=['individualID'])
df_left.head()
```

# Visual representation: INNER join
* Output Table will have only the matching rows from all tables

```python
path = '/Users/mhevia/gitHub2/testz/INNER-JOIN.png'
Image(filename=path, width=5000, height=500)
```

```python
df_i = pd.merge(df1, df2, how='inner', left_on=['individualID'], right_on=['individualID'])
df_i.head()
```

```python
df_ii = pd.merge(df3, df4, how='inner', left_on=['individualID'], right_on=['individualID'])
df_ii.head()
```

```python
df_inner = pd.merge(df_i, df_ii, how='outer', left_on=['individualID'], right_on=['individualID'])
df_inner.head()
```

# Visual representation: Outer join
* Output Table will have all rows from all tables, regardless if they match or not

```python
path = '/Users/mhevia/gitHub2/testz/FULL-OUTER-JOIN.png'
Image(filename=path, width=5000, height=500)
```

```python
df_o = pd.merge(df1, df2, how='outer', left_on=['individualID'], right_on=['individualID'])
df_o.head()
```

```python
df_oo = pd.merge(df3, df4, how='outer', left_on=['individualID'], right_on=['individualID'])
df_oo.head()
```

```python
df_outer = pd.merge(df_o, df_oo, how='outer', left_on=['individualID'], right_on=['individualID'])
df_outer.head()
```
